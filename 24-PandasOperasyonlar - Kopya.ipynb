## groupby

```python
maasSozlugu = {"Departman" : ["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
               "Çalışan İsmi" : ["Ahmet","Mehmet","Atıl","Burak","Zeynep","Fatma"],
              "Maas" : [100,150,200,300,400,500]}
```

```python
maasDataFrame = pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
grupObjesi =maasDataFrame.groupby("Departman")
```

```python
grupObjesi.mean()
```

```python
grupObjesi.count()
```

```python
grupObjesi.max()
```

```python
grupObjesi.min()
```

```python
grupObjesi.describe() ##istatistik değerleri veriyor
```
