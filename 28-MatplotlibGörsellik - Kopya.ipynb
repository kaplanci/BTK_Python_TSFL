```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
numpyDizisi1 = np. linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1**2
```

```python
numpyDizisi2
```

```python
(benimFigure,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
benimEksen.plot(numpyDizisi2,numpyDizisi1,"b")
```

```python
(yeniFigure , yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 2,color = "blue",linewidth = 4,marker = "+")   #linewidth çizgi kalınlığı ayarı
yeniEksen.plot(numpyDizisi1,numpyDizisi1 + 3,color = "yellow",linewidth = 7)

yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,color ="red",linestyle = "-.") 
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,color ="green",linestyle = "--") 
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,color ="pink",linestyle = ":") #linestyle çizgi modeli belirler
```
