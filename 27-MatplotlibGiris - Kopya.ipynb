```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline  
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
#matplotlibe geçircez np dizlileri
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g") ##ilk data x axis yazıldı ikinci data y axis,g ise yeşil olmasını sağladı
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")  #eksenler değişecekse yerler değişmeli


plt.xlabel("yas verisi")
plt.ylabel("Kilo verisi")
plt.title("Kilonun yaşa göre değişimi")

#plt.show() normalde yazılmalı fakat en başta inlien komutu ile bunun önüne geçtik
```

```python
(benimFigure ,benimEksenler )= plt.subplots(nrows = 1,ncols = 2) 
 #benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")


for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
    
plt.tight_layout()    
```

```python
yeniFigure =plt.figure(figsize=(4,4))

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label = "numpyDizisi**2")   #verinin açıklaması yapıldı
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label = "numpyDizisi**3")
yeniEksen.legend(loc = 2)  #konum verme  lejanttan hatırla
```

```python
yeniFigure =plt.figure(dpi=(150))  #dpi kalite verir

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label = "numpyDizisi**2")   #verinin açıklaması yapıldı
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label = "numpyDizisi**3")
yeniEksen.legend(loc = 2)  #konum verme  lejanttan hatırla
```
